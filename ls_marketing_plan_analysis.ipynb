In particular, this particular income bracket is 13.2% of Fort Collins' adult population; the next largest bracket is $100,000-$124,999 (9.57%), followed by $60,000-$74,999 (8.12%), $125,000-$149,999 (7.04%), $25,000-$29,999 (6.97%), $50,000-$59,999 (6.80%), and $150,000-$199,999 (6.48%). The income distribution of the city trends towards the upper-middle class and upper-class. Over 50% of earners in Ft. Collins make above $60k/year; for Larimer County, that number is closer to 55%. On the lower end of the income scale, 19.65% of Ft. Collins earners make between $25k-$44.9k. Data suggests an overall upward trend in median income across the county and in Fort Collins in particular.*

- *Fort Collins is the fourth most populous city in Colorado (est. 165k for 2019), after Denver, Colorado Springs, and Aurora, respectively. The city and surrounding area is predominately white, affluent, and educated. Home to Colorado State University-Fort Collins, the city attracts a young and politically progressive demographic, although the city's Democratic leanings are tempered by the more conservative surrounding area.*

- *While the city does have a minority population, the city is so overwhelmingly white that it makes sense to focus on constructing first avatars towards the ethnic majority of the population. Minority-focused strategies can be employed as the brand gains a broader footprint outside of Fort Collins.*

- *The top occupation for the area are the overwhelmingly female dominated administrative jobs (11.7%), followed by sales (11.6%) and management (10.9%), which lean male. The top part-time job is the slightly male-leaning food service (70%), followed by the female-dominated personal care jobs (67.4%).*

- *For the median income of full-time civilian employees in Fort Collins, legal has the highest ($93k/yr), followed by engineering ($81.5k/yr) and computers & math ($79.6k/yr). The median income for administrative jobs is $36.5k/yr, while sales is $40.4k/yr and management is $64.8k/yr. The median income for food service jobs is $20.5k/yr, and personal care is $22.7k/yr.*
